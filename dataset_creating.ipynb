```python
import os
import re
import pandas as pd
```

```python
path = 'dataset/'
path_neg = 'neg/'
path_neu = 'neu/'
path_pos = 'pos/'
```

```python
folder_path_neg = path + path_neg
file_names = os.listdir(folder_path_neg)
print(file_names)
```

```python
with open(folder_path_neg + file_names[0], 'r') as file:
    text = file.read()

text
```

```python
def cleaner(text_line):
    clean_text = re.sub(r'[^a-zA-Zа-яА-ЯёЁ0-9\.\,\!\?\:\;\"\'\(\)\-\s]', '', text_line)
    clean_text = re.sub(r'[\n\r\t\\]', '', clean_text)
    clean_text = clean_text.strip()
    return clean_text


cleaner(text)
```

```python
data = {
    'text': [],
    'label': [],
    'link': []
}

label = 0
for path_to_files in [path + path_neg,
                      path + path_neu,
                      path + path_pos]:
    
    spisok_file_names = os.listdir(path_to_files)

    for name_txt in spisok_file_names:
        with open(path_to_files + name_txt, 'r') as file:
            text = file.read()
        
        clean_text = cleaner(text)
        data['text'].append(clean_text)
        data['label'].append(label)
        data['link'].append(path_to_files + name_txt)
    
    label += 1
```

```python
data = pd.DataFrame(data)
data['label'].value_counts()
```
